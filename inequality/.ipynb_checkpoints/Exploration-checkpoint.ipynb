## GDP

```python
gdp = pd.read_csv(os.path.join(base_url, "data", "gdp2.csv"))
print(gdp.shape)
gdp.head(20)
```

```python
years = gdp.columns.to_list()
years.remove("Country Name")
years.remove("Country Code")
print(len(years))
print(years)
```

```python
gdp_col = []
name_col = []
year_col = []
code_col = []
for idx, row in gdp.iterrows():   
    for year in years:
        name_col.append(row["Country Name"])
        code_col.append(row["Country Code"])
        gdp_col.append(row[year])
        year_col.append(year)
```

```python
expanded_df = pd.DataFrame()
expanded_df["year"] = year_col
expanded_df["country"] = name_col
expanded_df["code"] = code_col
expanded_df["gdp"] = gdp_col
print(expanded_df.shape)
expanded_df.head(20)
```

```python
fig = px.choropleth(
    expanded_df,
    locations='code',
    locationmode='ISO-3',
    hover_name="country",
    color='gdp',
    range_color=(0,100000),
    color_continuous_scale="plasma",
    animation_frame='year'
)
fig.show()
```
